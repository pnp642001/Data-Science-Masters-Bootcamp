Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
data=sns.load_dataset('titanic')
data.head()
```

```python
import plotly.express as px
figure=px.scatter(data,x='age',y='fare')
figure.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
data=sns.load_dataset('tips')
data.head()
```

```python
figure_2=px.box(data,x='day',y='tip',color='smoker')
figure_2.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
figure_3=px.histogram(data,x='sex',y='total_bill',pattern_shape='smoker',color='day')
figure_3.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.

```python
data=sns.load_dataset('iris')
data.head()
```

```python
figure_4=px.scatter_matrix(data,dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],color='species')
figure_4.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.<br>
ANS:<br>
distplot is a function in the seaborn library that creates a histogram with a kernel density estimate (KDE) overlaid on top of it. It provides a convenient way to visualize the distribution of a single variable. However, Plotly express does not have a built-in distplot function. 

```python
data=sns.load_dataset('tips')
data.head()
```

```python
figure_5 = px.histogram(data, x='total_bill', color='sex', marginal='rug', opacity=0.7,histnorm='density', nbins=30)
figure_5.show()
```
