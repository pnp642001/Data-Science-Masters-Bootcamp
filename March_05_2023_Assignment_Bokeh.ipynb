```python
from bokeh.plotting import figure, output_file, show

# create some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# output to static HTML file
output_file("line_graph.html")

# create a new plot with a title and axis labels
p = figure(title="Line Plot", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
p.line(x, y, legend_label="Line", line_width=2)

# show the results
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.<br>
ANS:<br>
Glyphs in Bokeh are visual shapes or markers used to represent data points in a plot. Bokeh provides a wide range of glyph types such as circles, squares, lines, and more. Glyphs can be added to a Bokeh plot using the add_glyph() method of a Figure object.

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource, Circle

# create some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# create a ColumnDataSource with the data
source = ColumnDataSource(data=dict(x=x, y=y))

# output to static HTML file
output_file("glyphs.html")

# create a new plot with a title and axis labels
p = figure(title="Glyphs Example", x_axis_label='x', y_axis_label='y')

# create a Circle glyph instance with the size and fill color
circle_glyph = Circle(x='x', y='y', size=10, fill_color="blue")

# add the Circle glyph instance to the plot using the data from the source
p.add_glyph(source, circle_glyph)

# show the results
show(p)

```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?<br>
ANS:<br>
Bokeh provides several ways to customize the appearance of a plot, including the axes, title, and legend. Here are some common ways to do this:

Customize the axes:

* To set the label of the x-axis, use the xaxis.axis_label attribute of the Figure object. For example, p.xaxis.axis_label = "X Label"

* To set the label of the y-axis, use the yaxis.axis_label attribute of the Figure object. For example, p.yaxis.axis_label = "Y Label"

* To set the range of the x-axis, use the x_range attribute of the Figure object. For example, p.x_range = Range1d(0, 10)
 
* To set the range of the y-axis, use the y_range attribute of the Figure object. For example, p.y_range = Range1d(0, 20)

* To customize the ticks and labels of the x-axis or y-axis, use the Axis class. For example, p.xaxis.ticker = [0, 5, 10] sets the x-axis tick locations to 0, 5, and 10.

Customize the title:

* To set the title of the plot, use the title attribute of the Figure object. For example, p.title.text = "My Plot"

* To customize the appearance of the title, such as changing the font size or color, use the title property of the Title class. For example, p.title.text_font_size = "20pt" sets the font size of the title to 20 points.

Customize the legend:

* To add a legend to the plot, use the legend attribute of the Figure object. For example, p.legend.label_text_font_size = "14pt" adds a legend and sets the font size of the legend text to 14 points.

* To set the location of the legend, use the legend.location attribute. For example, p.legend.location = "top_right" sets the legend location to the top-right corner of the plot.

* To customize the appearance of the legend, use the Legend class. For example, p.legend.background_fill_alpha = 0.5 sets the transparency of the legend background to 50%.

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?<br>
ANS:<br>
Bokeh server is a Python library that allows you to create interactive web-based visualizations and applications with Bokeh. It enables the creation of dynamic and interactive plots that can be updated in real-time based on user interactions or changes to the underlying data. The Bokeh server provides a flexible architecture that allows for building complex and scalable applications with ease.

To use the Bokeh server, you first need to define a Bokeh application that defines the layout and behavior of the interactive plot. The application can include one or more plots and widgets, which can be used to control the behavior of the plots. The Bokeh server then serves the application as a web page that can be accessed by users in a web browser.


Here are the basic steps to create a Bokeh server application:

1. Define the data and create the initial plot(s) that will be displayed.

2. Define any widgets that will be used to control the behavior of the plot(s).

3. Define a callback function that will be called whenever the user interacts with a widget.

4. Register the callback function with the widget(s) so that the plot(s) can be updated in real-time based on the user's input.

5. Create a curdoc() object that represents the current Bokeh document.

6. Add the plot(s) and widget(s) to the document.

7. Start the Bokeh server by running bokeh serve from the command line.
