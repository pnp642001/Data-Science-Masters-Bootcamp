Q1. What is Flask Framework? What are the advantages of Flask Framework?<br>
ANS:<br>
Flask is a lightweight and flexible web application framework for Python. It is a micro-framework that provides developers with the essential tools and features needed to build web applications quickly and efficiently. Some of the key features of Flask include:

Routing: Flask allows developers to define URL routes for their applications, making it easy to map incoming requests to the appropriate functions.

Templating: Flask supports a range of templating engines, making it easy to create dynamic HTML pages.

Database Integration: Flask can be easily integrated with a wide range of databases, including SQLite, MySQL, and PostgreSQL.

Flask Extensions: Flask has a large community of developers who have built extensions that add new functionality to the framework, such as Flask-SQLAlchemy for database integration and Flask-Login for user authentication.

Some advantages of using Flask include:

Lightweight: Flask is a micro-framework that does not require much overhead, making it a great choice for small to medium-sized applications.

Flexibility: Flask allows developers to choose the components they need for their application, making it easy to build a custom solution that meets their specific requirements.

Easy to Learn: Flask has a simple and intuitive syntax that makes it easy to get started with web development in Python.

Community Support: Flask has a large and active community of developers who contribute to the framework and build extensions that add new functionality to it.

Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?<br>
ANS:<br>
In Flask, App routing refers to the process of defining URL routes for your web application. When a user makes a request to your Flask application, Flask uses the URL route defined in your application to determine which view function should handle the request.

App routes are essential in Flask because they allow you to map incoming requests to the appropriate functions that generate a response. By defining app routes, you can create a clean and logical structure for your application, making it easier to maintain and update.

Overall, app routing is an essential part of building Flask applications, and it allows you to create a clean and organized structure for your application while providing a simple and intuitive interface for users to interact with your application.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route("/Welcome")
def welcome_message():
    return "Welcome to ABC Corporation"

@app.route("/")
def company_details():
    return "<h2>Company Name: ABC Corporation</h2><h2>Location: India</h2><h2>Contact Detail: 999-999-9999</h2>"
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.<br>
ANS:<br>
The url_for() function is used in Flask for URL building. This function generates a URL to the specified endpoint with the method provided.

Here is a simple example of how to use url_for() function in Python Flask:<br>


```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

@app.route('/show_url')
def show_url():
    index_url = url_for('index')
    profile_url = url_for('profile', username='John')
    return f'Index URL: {index_url}, Profile URL: {profile_url}'




if __name__=="__main__":
    app.run(host="0.0.0.0")
```
