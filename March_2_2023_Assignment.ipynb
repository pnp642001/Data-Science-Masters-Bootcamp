Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.<br>
ANS:<br>
A scatter plot is a type of plot or mathematical diagram that is used to display values for two variables for a set of data. It is a graph in which the values of two variables are plotted along two axes, with one variable on the x-axis and the other variable on the y-axis. Each data point is represented as a dot or a circle on the plot, and the position of the dot or circle indicates the values of the two variables. Scatter plots are commonly used to visualize the relationship between two variables and to identify patterns or trends in the data.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.<br>
ANS:<br>
The subplot() function in Matplotlib is used to create multiple plots within a single figure. It takes three arguments: the number of rows of subplots, the number of columns of subplots, and the index of the subplot to create.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a main title for the entire figure
fig.suptitle('Four Line Plots')

# Show the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.<br>
ANS:<br>
A bar plot is a type of plot used to display and compare data in a categorical or discrete dataset. It represents each category as a rectangular bar with height proportional to the values being plotted. The bars can be displayed vertically or horizontally depending on the nature of the data and the purpose of the plot.

Bar plots are used to show the distribution, comparison, and trends of data across different categories or groups. They are commonly used to compare data sets, track changes over time, and identify trends and patterns in the data. Bar plots are often used in fields such as business, economics, finance, and social sciences to analyze and visualize data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color='Red')
plt.title('Vertical Bar Plot')
plt.show()
```

```python
plt.barh(company,profit,color="blue")
plt.title("horizontal bar plot")
plt.show()
```

Q5: What is a box plot? Why is it used?<br>
ANS:<br>
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of the data: minimum, first quartile, median, third quartile, and maximum. The box plot is used to identify outliers and to compare the distribution of multiple datasets.

The box plot consists of a box that represents the second and third quartiles, with a vertical line inside the box representing the median. The whiskers extend from the box to the minimum and maximum values, excluding outliers, which are represented as individual points. Outliers are defined as data points that are more than 1.5 times the interquartile range (IQR) away from the upper or lower quartile.

Box plots can be useful in a variety of fields, such as statistics, finance, and medicine, to visualize and compare the distribution of data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data=[box1,box2]
plt.boxplot(data)
plt.title("box plot")
plt.show()
```
